```
import numpy as np
import math
import random

import time
#-----------------------------------#
#----------Shor's Algorism----------#
#-----------------------------------#
def shor(N):
  #-----Step 1-----#
  if N == 2:print
  if is_even(N):
    print_prime_factor(N, 2)
    return True

  #-----Step 2-----#
  TF, a, b = is_perfect_power(N)
  if TF:
    print('{} = {}^{}' .format(N, a, b))
    print_prime_factor(N, a)
    return True

  #-----Step 3, 4, 5-----#
  NumberList = []
  for i in range(1, N):
    NumberList.append(i)
  
  while 1:
    x = pick_x(NumberList)
    print('I pick {}.'.format(x))
    NumberList.remove(x)
    r = order_calculation(N, x)
    print('r = {}'.format(r))
    if is_even(r) and x ** int(r / 2) % N != -1:
      a, b = generate_factor(N, x, r)
      if a != 1 and N != a:
        print_prime_factor(N, a)
        return 0
      if b != 1 and N != b:
        print_prime_factor(N, b)
        return 0
    print('Failed.')
    print('I pick x again.')
    print(NumberList)
    if not NumberList:
      print('I had picked all x.')
      break

  print('{} is prime number!' .format(N))
  return 0

#-------Even_Judgment
def is_even(M):
  return M % 2 ==0


#-------Perfect_Power_Judgment
def is_perfect_power(N):
  size = int(math.log2(N) + 1)
  for b in range(2, size):
    a = int(math.pow(N, 1 / b))
    if a ** b == N or (a + 1) ** b == N:
      return True, a, b
  return False, 0, 0

#-------factor
def generate_factor(N, x, r):
  return gcd(N, x**int(r / 2) - 1), gcd(N, x**int(r / 2) + 1)

#-------Pick_x
def pick_x(NumberList):
  return random.choice(NumberList)

#-------Print_Prime_Factor
def print_prime_factor(N, a):
  print('{} has a prime factor of {}' .format(N, a))
  return 0

#-------Greatest Common Divisor
def gcd(x, y):
  if y == 0:
    return x
  else:
    return gcd(y, x % y)

#-------Order_Calculation
def order_calculation(N, x):
  while 1:
    i = x ** random.randrange(1, N + 1) % N
    r = 0
    r1 = 0
    for j in range(1, N + 1):
      if x ** j % N == i:
        if r1 == 0:
          r1 = j
        else:
          r = j - r1
          return r

#------------------------------------#
#------------MAIN PROGRAM------------#
#------------------------------------#
N = int(input('N = '))
if N<2:
  while N<2:
    print('Error! N should be more than 1')
    N = int(input('N = '))
start = time.time()
a = shor(N)
process_time = time.time() - start
print('Process Time')
print(process_time)
```
